# IEEE SSCS AI Sub-Team      Data Visualization Task
This notebook covers all 6 task sections using Matplotlib and Seaborn with the Iris, Tips, Flights, and custom datasets.

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns

# Load datasets
iris    = sns.load_dataset('iris')
tips    = sns.load_dataset('tips')
flights = sns.load_dataset('flights')

print('iris   :', iris.shape)
print('tips   :', tips.shape)
print('flights:', flights.shape)
```

```python
# 1a: Line plot  total passengers per year (Flights) 
yearly = flights.groupby('year')['passengers'].sum().reset_index()

fig, ax = plt.subplots(figsize=(9, 4))
ax.plot(
    yearly['year'], yearly['passengers'],
    marker='o', color='steelblue', linewidth=2, markersize=6
)
ax.set_title('Total Annual Passengers (1949–1960)', fontsize=14, fontweight='bold')
ax.set_xlabel('Year')
ax.set_ylabel('Total Passengers')
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{int(x):,}'))
ax.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
#  1b: Scatter plot  sepal length vs width, coloured by species (Iris) 
palette = {'setosa': '#E63946', 'versicolor': '#2A9D8F', 'virginica': '#E9C46A'}

fig, ax = plt.subplots(figsize=(7, 5))
for species, group in iris.groupby('species'):
    ax.scatter(
        group['sepal_length'], group['sepal_width'],
        label=species.capitalize(), color=palette[species],
        alpha=0.75, edgecolors='white', linewidths=0.5, s=70
    )
ax.set_title('Sepal Length vs Width by Species', fontsize=14, fontweight='bold')
ax.set_xlabel('Sepal Length (cm)')
ax.set_ylabel('Sepal Width (cm)')
ax.legend(title='Species')
ax.grid(linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
#  1c: Seaborn scatter  bill vs tip (Tips) 
fig, ax = plt.subplots(figsize=(7, 5))
sns.scatterplot(
    data=tips, x='total_bill', y='tip',
    hue='time', style='sex', palette='Set2', alpha=0.8, ax=ax
)
ax.set_title('Total Bill vs Tip Amount', fontsize=14, fontweight='bold')
ax.set_xlabel('Total Bill ($)')
ax.set_ylabel('Tip ($)')
ax.grid(linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

## Step 2  Styling & Customization

```python
# 2a: Styled multi-line plot with theme (Flights) 
pivot = flights.pivot(index='month', columns='year', values='passengers')
months_order = [
    'January','February','March','April','May','June',
    'July','August','September','October','November','December'
]
pivot = pivot.reindex(months_order)

with plt.style.context('seaborn-v0_8-whitegrid'):
    fig, ax = plt.subplots(figsize=(11, 5))
    cmap = plt.get_cmap('Blues')
    years = pivot.columns.tolist()
    for i, year in enumerate(years):
        color = cmap(0.3 + 0.7 * i / len(years))
        ax.plot(pivot.index, pivot[year], color=color, linewidth=1.6, label=str(year))
    ax.set_title('Monthly Passengers per Year — Styled', fontsize=14, fontweight='bold')
    ax.set_xlabel('Month')
    ax.set_ylabel('Passengers')
    ax.tick_params(axis='x', rotation=45)
    ax.legend(title='Year', bbox_to_anchor=(1.01, 1), loc='upper left', fontsize=8)
    plt.tight_layout()
    plt.show()
```

```python
#  2b: Custom palette & annotation (Tips) 
fig, ax = plt.subplots(figsize=(7, 5))
sns.scatterplot(
    data=tips, x='total_bill', y='tip',
    hue='smoker', palette={'Yes': '#E76F51', 'No': '#264653'},
    alpha=0.85, s=60, ax=ax
)
# Add regression line manually
m, b = np.polyfit(tips['total_bill'], tips['tip'], 1)
x_line = np.linspace(tips['total_bill'].min(), tips['total_bill'].max(), 100)
ax.plot(x_line, m * x_line + b, color='black', linewidth=1.5, linestyle='--', label='Trend')
ax.annotate(
    f'slope = {m:.3f}',
    xy=(35, m * 35 + b), xytext=(28, 8),
    arrowprops=dict(arrowstyle='->', color='black'),
    fontsize=9
)
ax.set_title('Tip vs Bill — Smoker vs Non-Smoker', fontsize=13, fontweight='bold')
ax.set_xlabel('Total Bill ($)')
ax.set_ylabel('Tip ($)')
ax.legend(title='Smoker')
ax.grid(linestyle=':', alpha=0.4)
plt.tight_layout()
plt.show()
```

## Step 3  Distribution Plots

```python
# 3a: Histogram + KDE  tip amount (Tips) 
fig, ax = plt.subplots(figsize=(8, 4))
sns.histplot(tips['tip'], bins=20, kde=True, color='steelblue', ax=ax)
ax.axvline(tips['tip'].mean(), color='red', linestyle='--', linewidth=1.5, label=f"Mean: ${tips['tip'].mean():.2f}")
ax.axvline(tips['tip'].median(), color='orange', linestyle='--', linewidth=1.5, label=f"Median: ${tips['tip'].median():.2f}")
ax.set_title('Distribution of Tip Amounts', fontsize=13, fontweight='bold')
ax.set_xlabel('Tip ($)')
ax.set_ylabel('Count')
ax.legend()
plt.tight_layout()
plt.show()
```

```python
# 3b: KDE plot petal length by species (Iris) 
fig, ax = plt.subplots(figsize=(8, 4))
for species in iris['species'].unique():
    subset = iris[iris['species'] == species]['petal_length']
    sns.kdeplot(subset, ax=ax, label=species.capitalize(), fill=True, alpha=0.35)
ax.set_title('KDE of Petal Length by Species', fontsize=13, fontweight='bold')
ax.set_xlabel('Petal Length (cm)')
ax.set_ylabel('Density')
ax.legend(title='Species')
plt.tight_layout()
plt.show()
```

```python
# 3c: ECDF plot  sepal length by species (Iris) 
fig, ax = plt.subplots(figsize=(8, 4))
sns.ecdfplot(data=iris, x='sepal_length', hue='species', ax=ax)
ax.set_title('ECDF of Sepal Length by Species', fontsize=13, fontweight='bold')
ax.set_xlabel('Sepal Length (cm)')
ax.set_ylabel('Cumulative Proportion')
ax.grid(linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

## Step 4 Categorical & Comparison Plots

```python
# 4a: Bar chart  average tip by day and time (Tips) 
avg_tip = tips.groupby(['day', 'time'])['tip'].mean().reset_index()
day_order = ['Thur', 'Fri', 'Sat', 'Sun']

fig, ax = plt.subplots(figsize=(8, 4))
sns.barplot(
    data=avg_tip, x='day', y='tip', hue='time',
    order=day_order, palette='Set2', ax=ax
)
ax.set_title('Average Tip by Day and Meal Time', fontsize=13, fontweight='bold')
ax.set_xlabel('Day of Week')
ax.set_ylabel('Average Tip ($)')
ax.legend(title='Time')
ax.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# 4b: Box plot  total bill by day (Tips) 
fig, ax = plt.subplots(figsize=(8, 4))
sns.boxplot(
    data=tips, x='day', y='total_bill', hue='day',
    order=day_order, palette='pastel', legend=False,
    flierprops=dict(marker='x', markeredgecolor='red'), ax=ax
)
ax.set_title('Total Bill Distribution by Day', fontsize=13, fontweight='bold')
ax.set_xlabel('Day')
ax.set_ylabel('Total Bill ($)')
ax.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
# 4c: Violin plot  petal width by species (Iris) 
fig, ax = plt.subplots(figsize=(7, 5))
sns.violinplot(
    data=iris, x='species', y='petal_width', hue='species',
    palette='muted', inner='quartile', legend=False, ax=ax
)
ax.set_title('Petal Width Distribution by Species', fontsize=13, fontweight='bold')
ax.set_xlabel('Species')
ax.set_ylabel('Petal Width (cm)')
ax.grid(axis='y', linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

## Step 5  Heatmaps & Subplots

```python
#  5a: Heatmap  monthly passengers pivot (Flights) 
pivot_heat = flights.pivot(index='month', columns='year', values='passengers').reindex(months_order)

fig, ax = plt.subplots(figsize=(12, 6))
sns.heatmap(
    pivot_heat, annot=True, fmt='d', cmap='YlOrRd',
    linewidths=0.4, linecolor='white', ax=ax
)
ax.set_title('Monthly Airline Passengers 1949–1960', fontsize=14, fontweight='bold')
ax.set_xlabel('Year')
ax.set_ylabel('Month')
plt.tight_layout()
plt.show()
```

```python
# 5b: Correlation heatmap (Iris numeric features) 
corr = iris.drop(columns='species').corr()

fig, ax = plt.subplots(figsize=(6, 5))
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(
    corr, mask=mask, annot=True, fmt='.2f',
    cmap='coolwarm', center=0, linewidths=0.5, ax=ax
)
ax.set_title('Iris Feature Correlation Matrix', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# 5c: 2×2 Subplot grid (OOP API) 
fig, axes = plt.subplots(2, 2, figsize=(12, 9))
fig.suptitle('Tips Dataset — Multi-View Dashboard', fontsize=15, fontweight='bold', y=1.01)

# Top-left: histogram of total bill
sns.histplot(tips['total_bill'], bins=18, kde=True, color='steelblue', ax=axes[0, 0])
axes[0, 0].set_title('Total Bill Distribution')
axes[0, 0].set_xlabel('Total Bill ($)')

# Top-right: count of diners by day
sns.countplot(data=tips, x='day', hue='day', order=day_order, palette='Set3', legend=False, ax=axes[0, 1])
axes[0, 1].set_title('Number of Diners by Day')
axes[0, 1].set_xlabel('Day')

# Bottom-left: box plot of tip by smoker
sns.boxplot(data=tips, x='smoker', y='tip', hue='smoker', palette='pastel', legend=False, ax=axes[1, 0])
axes[1, 0].set_title('Tip Amount — Smoker vs Non-Smoker')
axes[1, 0].set_xlabel('Smoker')

# Bottom-right: violin of total bill by sex
sns.violinplot(data=tips, x='sex', y='total_bill', hue='sex', palette='muted', inner='quartile', legend=False, ax=axes[1, 1])
axes[1, 1].set_title('Bill Distribution by Gender')
axes[1, 1].set_xlabel('Gender')

plt.tight_layout()
plt.show()
```

## Step 6   Insight & Storytelling

### Iris Dataset
1. **Setosa is clearly separable** from the other two species in both petal length and petal width, making it the easiest class to classify.
2. **Petal dimensions are highly correlated** (r ≈ 0.96), while sepal width is nearly uncorrelated with the other three features   suggesting dimensionality reduction would work well here.
3. **Virginica has the largest spread** in petal width, implying more within-species variation than versicolor.

### Tips Dataset
1. **Tip amount grows with bill size** (positive linear trend, slope ≈ 0.105), but tipping percentage is not constant   large bills do not always yield proportionally larger tips.
2. **Saturday sees the most diners and the highest variance in bills**, making it the most economically significant day for the restaurant.
3. **Smokers show slightly higher median tips** but also more outliers, suggesting a wider spread of tipping behaviour in that group.

### Flights Dataset
1. **Passenger counts grow every year** from 1949 to 1960   the trend is clear across all months without exception.
2. **July and August are consistently the peak months** in every year, revealing a strong and stable seasonal pattern tied to summer holidays.
3. **The absolute gap between summer peaks and winter troughs widens over time**, meaning seasonality amplifies as the industry grows   not just the baseline.
